```python
from diffusers import StableDiffusionPipeline
import torch
```

```python
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

```python
prompt = "A futuristic AI robot teaching students in a modern classroom, digital art"
```

```python
image = pipe(prompt).images[0]
display(image)
```

```python
prompts = [
    "A futuristic smart city with flying cars, digital art",
    "An AI robot helping doctors in a hospital",
    "A beautiful sunset over mountains in fantasy style"
]

for p in prompts:
    image = pipe(p).images[0]
    display(image)
```
